```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### importing the dataset

```python
dataset = pd.read_excel(r"C:\Users\namra\Documents\auto-mpg.xlsx")
dataset
```

```python
df = dataset.drop(['mpg' , 'carname' , 'carname2'], axis =1)
df.rename(columns = {'mpg2':'mpg'}, inplace = True)
df.rename(columns = {'carname3':'carname'}, inplace = True)
df
```

```python
df.info()
```

### checking for null values

```python
df.dropna()
```

```python
df.horsepower.unique()
```

```python
df['horsepower']=df['horsepower'].replace('?          ',np.nan)
df['horsepower'].isnull().sum()
```

```python
df['horsepower']=df['horsepower'].astype('float64')
# fill the nan values with median values
df['horsepower']=df['horsepower'].fillna(df['horsepower'].median())
```

### checking correaltion between attributes

```python
corr = df.corr()
plt.figure(figsize=(20,10))
g=sns.heatmap(corr,annot=True)
g.set(title='Correlation matrix of the car-mpg dataset')
plt.show()  

#Mpg column has strong correlation on cylinders,displacement,weight
```

```python
df['cylinders'].value_counts()
```

```python
import statsmodels.formula.api as smf
test=smf.ols('mpg~cylinders+displacement+horsepower+weight+acceleration+origin',df).fit()
test.summary()
```

### removing irrelevant attributes

```python
df1 = df.drop(['carname' , 'acceleration', 'origin'],axis=1)
df1
```

### Training the model

```python
y = df1['mpg']
x = df1.drop(['mpg'] , axis=1)
x
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm
```

```python
lm.fit(X_train, y_train)
```

```python
print(lm.intercept_)
print(lm.coef_)
```

**mpg** = -18.026 + 2.36x**cylinders** - 7.64x10^(-3)x**displacement** + 2.065x10^(-4)x**horsepower** - 6.35x10^(-3)x**weight** + 7.97x10^(-1)x**modelyear**

### model accuracy

```python
lm.score(X_test, y_test)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, lm.predict(X_test))
```

### making predictions

```python
#mpg = -18.026 + 2.36xcylinders - 7.64x10^(-3)xdisplacement + 2.065x10^(-4)xhorsepower - 6.35x10^(-3)xweight + 7.97x10^(-1)xmodelyear
def predict_y(cyl, disp, hp, wt, year):
    mpg = -18.026 + 2.36*cyl - (7.64*(10**(-3)))*disp + (2.065*(10**(-4)))*hp - (6.35*(10**(-3)))*wt + (7.97*(10**(-1)))*year
    return mpg
```
